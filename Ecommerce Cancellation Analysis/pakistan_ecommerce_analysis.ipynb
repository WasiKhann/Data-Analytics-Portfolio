### Overview of the Dataset

This dataset represents the largest collection of retail e-commerce orders from Pakistan available to the public. It encompasses half a million transaction records spanning from March 2016 to August 2018, sourced from diverse e-commerce merchants as part of a research initiative.

### Dataset Contents

**Geography:** Pakistan  
**Time Period:** March 2016 â€“ August 2018  
**Unit of Analysis:** E-Commerce Orders  

The dataset comprises comprehensive details on half a million e-commerce transactions in Pakistan, covering item specifics, shipping and payment methods (including credit card, Easy-Paisa, Jazz-Cash, and cash-on-delivery), product categories (such as fashion, mobile, electronics, appliances, etc.), order dates, SKUs, prices, quantities, totals, and customer IDs. This dataset stands as the most exhaustive publicly available resource for understanding the landscape of e-commerce in Pakistan.

**Data Description:**

- **item_id:** Identifier for each item in the dataset.
- **status:** Status of the order (e.g., completed, cancelled, refunded).
- **created_at:** Date and time when the order was created.
- **sku:** Stock Keeping Unit, a unique identifier for each product.
- **price:** Unit price of the item.
- **qty_ordered:** Quantity of items ordered in each transaction.
- **grand_total:** Total amount of the transaction.
- **increment_id:** Incremental identifier for each order.
- **category_name_1:** Category name of the product.
- **sales_commission_code:** Code representing sales commission (if applicable).
- **discount_amount:** Amount of discount applied to the transaction.
- **payment_method:** Method used for payment (e.g., credit card, cash on delivery).
- **Working Date:** Date of the transaction.
- **BI Status:** Business intelligence status (if applicable).
- **MV:** Market Value.
- **Year:** Year of the transaction.
- **Month:** Month of the transaction.
- **Customer Since:** Date since when the customer has been associated with the platform.
- **M-Y:** Month and Year combined.
- **FY:** Fiscal Year.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', 100)

# Load the dataset
df = pd.read_csv('Pakistan Largest Ecommerce Dataset.csv')

# Display initial data information
print(f"Data consists of {df.shape[0]} rows and {df.shape[1]} columns")
df.info()

# Handling missing values
print(df.isnull().sum())

# Remove columns with all null values
columns_to_remove = ['Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25']
df.drop(columns=columns_to_remove, axis=1, inplace=True)

# Remove rows where all the records are empty
df.dropna(how='all', inplace=True)

# Visualize remaining missing values
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title('Missing Values Heatmap')
plt.show()

# Impute missing values with the most frequent values for certain columns
for column in ['sales_commission_code', 'category_name_1', 'sku', 'status', 'Customer ID', 'Customer Since']:
    df[column] = df[column].fillna(df[column].mode()[0])

# Check missing values after imputation
print(df.isnull().sum().sort_values(ascending=False))

# Save the cleaned dataset to a new CSV file
df.to_csv('Cleaned_Pakistan_Ecommerce_Dataset.csv', index=False)
```
